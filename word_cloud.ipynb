```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
import csv
```

```python
dissertation = open('files/dissertacaonatali.tex', 'r').read()
```

```python
stopwords = []
f = open('files/stopwords.txt', 'r')
g = open('files/stopwords_latex.txt', 'r')
h = open('files/stopwords_specific.txt', 'r')
for row in f:
   stopwords.append(row.replace('\n', ''))
for row in g:
   stopwords.append(row.replace('\n', ''))
for row in h:
   stopwords.append(row.replace('\n', ''))
```

```python
#Saving a mask, a black hole mask, to the word cloud
BH_mask = np.array(Image.open('figures/black_hole.png'))
```

```python
#Ininiating the word cloud object
diss_natali = WordCloud(background_color = 'white',stopwords = stopwords, mask = BH_mask, colormap = 'inferno')
#You can change your colormap as you want. See https://matplotlib.org/users/colormaps.html for more details
```

```python
#Generating the word cloud
diss_natali.generate(dissertation)

#Displaying the cloud
fig = plt.figure()
fig.set_figwidth(20) # set width
fig.set_figheight(20) # set height

plt.imshow(diss_natali, interpolation='bilinear')
plt.axis('off')
plt.savefig('figures/natali_dissertation-word_cloud.png')
```
