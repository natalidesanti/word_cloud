# Word cloud notebook

In this notebook you will find an quick way to transform any `tex` file into a `word cloud`

The examples you can see here correspond to my Master's and PhD thesis converted into it,
but you can easily upload your own file and work with it!

Let's first import the necessary libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
import csv
```

Now it is time to import the contents:

```python
dissertation = open('files/dissertacaonatali.tex', 'r').read()
thesis = open('files/thesisnatali.tex', 'r').read()
```

The challenge behind this task is making the stopwords files

This means you will have to check (or use my examples and improve them) what words are not supposed to be used in the word cloud you wanna create

Notice I have created a file with `latex` usual commands and other with the words in the different languages

```python
stopwords_portuguese = []
f = open('files/stopwords-portuguese.txt', 'r')
g = open('files/stopwords_latex.txt', 'r')
for row in f:
   stopwords_portuguese.append(row.replace('\n', ''))
for row in g:
   stopwords_portuguese.append(row.replace('\n', ''))
```

```python
stopwords_english = []
f = open('files/stopwords-english.txt', 'r')
g = open('files/stopwords_latex.txt', 'r')
for row in f:
   stopwords_english.append(row.replace('\n', ''))
for row in g:
   stopwords_english.append(row.replace('\n', ''))
```

Your **word cloud** can be created with any image you think represents better the context you have

Remember that choosing an image with a good contrast is essential to end up a good result!

Let's load the images:

```python
BH_mask = np.array(Image.open('figures/black_hole.png'))
galaxy_mask = np.array(Image.open('figures/galaxy.png'))
```

With all the ingredients we only need to define the **word clouds**

Here you can choose the `background color` and `colormap` you prefer:

```python
diss_natali = WordCloud(background_color = 'white', stopwords = stopwords_portuguese, mask = BH_mask, colormap = 'inferno')
thesis_natali = WordCloud(background_color = 'white', stopwords = stopwords_english, mask = galaxy_mask, colormap = 'inferno')
```

An finally create them:

```python
diss_natali.generate(dissertation)
thesis_natali.generate(thesis)
```

The final products can be seen in usual plots:

```python
#Masters
fig = plt.figure(dpi = 100, figsize = (10, 10))
plt.imshow(diss_natali, interpolation='bilinear')
plt.axis('off')
plt.savefig('figures/natali_dissertation-word_cloud.png')
```

For texts in portuguese, even including some words in the `stopwords` files, the final product is still not perfect

But you can see I have worked with **black holes**!

```python
fig = plt.figure(dpi = 100, figsize = (10, 10))
plt.imshow(thesis_natali, interpolation='bilinear')
plt.axis('off')
plt.savefig('figures/natali_thesis-word_cloud.png')
```
